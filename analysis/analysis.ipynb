```python
import pandas
import itertools
```

```python
raw = pandas.read_table('out', sep=' *')
raw.head()
```

```python
files = raw['file'].unique()
df = raw.pivot('s', 'file')
```

## Energy, magnetization, overlap

Plot the energy as a function of $s$ for each graph.

```python
df['E0'].plot(); ylabel('$E_0$');
df['m2'].plot(); ylabel('$m^2$');
df['q'].plot(); ylabel('$q$');
```

Plot differences as a function of $s$:

```python
for quant, label in [('E0', 'E_0'), ('m2', 'm^2'), ('q', 'q')]:
    figure()
    dfq = df[quant]
    for g1, g2 in itertools.combinations(files[:2], 2):
        abs(dfq[g1] - dfq[g2]).plot(label='%s, %s' % (g1, g2))
    ylabel('$\Delta %s$' % label)
    legend(loc='best');
```
