```python
import pandas
from itertools import combinations
```

```python
df = pandas.read_table('out', sep=' *', index_col=['file', 'h0', 's'])
graphs = df.index.get_level_values('file').unique()
df.head()
```

## Energy, magnetization, overlap for $h_0=1$

Plot measurements as a function of $s$.

```python
quants = ['energy', 'q2', 'mz2', 'mx2']
symbols = ['E_0', 'Q_2', 'm_z^2', 'm_x^2']
labels = ['${}$'.format(s) for s in symbols]

dfx = df.unstack('file')
dfx = dfx.ix[1.0] # select h_0 = 1

for quant, label in zip(quants, labels):
    dfx[quant].plot()
    ylabel(label)
```

Plot fractional differences as a function of $s$:

```python
dfx = df.xs(1.0, level='h0')
    
for ga, gb in combinations(graphs, 2):
    diffs = dfx.ix[ga] - dfx.ix[gb]
    sums  = dfx.ix[ga] + dfx.ix[gb]
    delta = (2. * diffs / sums).abs()
    delta.drop('iterations', 1).plot(logy=True)
    ylabel(r'$\Delta$')
    title('{}, {}'.format(ga, gb))
    #savefig('diff_h1_{}_{}.pdf')
```

# Overlap for different values of $h_0$

```python
dfx = df['q2'].unstack('h0')

for ga, gb in combinations(graphs, 2):
    diffs = dfx.ix[ga] - dfx.ix[gb]
    sums  = dfx.ix[ga] + dfx.ix[gb]
    delta = (2. * diffs / sums).abs()
    delta.plot()
    legend(title='$h_0$', loc='best')
    ylabel('$\Delta(Q_2)$')
    #savefig('diff_{}_{}.pdf')
```
