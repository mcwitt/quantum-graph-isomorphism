```python
import pandas
import itertools
```

```python
raw = pandas.read_table('out', sep=' *', skipinitialspace=True)
raw.head()
```

```python
files = raw['file'].unique()
df = raw.pivot('s', 'file')
```

## Energy, magnetization, overlap

Plot the energy as a function of $s$ for each graph.

```python
df['E0'].plot(); ylabel('$E_0$');
df['mz2'].plot(); ylabel('$m_z^2$');
df['mx2'].plot(); ylabel('$m_x^2$');
df['q'].plot(); ylabel('$q$');
```

Plot differences as a function of $s$:

```python
for quant, label in [('E0', 'E_0'), ('mz2', 'm_z^2'), ('mx2', 'm_x^2'), ('q', 'q')]:
    figure()
    dfq = df[quant]
    for g1, g2 in itertools.combinations(files[:2], 2):
        abs(2*(dfq[g1] - dfq[g2]) / (dfq[g1] + dfq[g2])).plot(label='{}, {}'.format(g1, g2))
    ylabel('$\Delta ({})$'.format(label))
    legend(loc='best');
```
