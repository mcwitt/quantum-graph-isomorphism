```python
import pandas
from itertools import combinations
```

```python
files = ['out-cg-n14', 'out-py-n14']
dfs = []

for filename in files:
    dfs.append(pandas.read_table(filename, sep=' *', index_col=['file', 'h0', 's']))
```

```python
df = pandas.read_table('out-cg-n14', sep=' *', index_col=['file', 'h0', 's'])
graphs = df.index.get_level_values('file').unique()
df.head()
```

```python
h0vals = df.index.get_level_values('h0').unique()
h0 = h0vals[h0vals.searchsorted(1)]
```

```python
dfx = df.xs(h0, level='h0').unstack('file')

for meas in dfx.columns.get_level_values(0).unique():
    figure()
    dfx[meas].plot()
    ylabel(meas)
```

```python
dfx = df['energy'].unstack('h0')

for ga, gb in combinations(graphs, 2):
    diffs = dfx.ix[ga] - dfx.ix[gb]
    sums  = dfx.ix[ga] + dfx.ix[gb]
    delta = (2. * diffs / sums).abs()
    delta[h0].plot(loglog=True)
    legend(title='$h_0$', loc='best')
    ylabel('$\Delta(Q_2)$')
    savefig('diff_q2.pdf')
```
