```python
import pandas
from itertools import combinations
```

```python
symbols = {'energy': '$E_0$',
           'mz2'   : '$m_z^2$',
           'mx2'   : '$m_x^2$',
           'q2'    : '$Q_2$'}

def symbol(s): return symbols[s] if s in symbols else s
```

```python
dfpy = pandas.read_csv('out-py-n14')
dfcg = pandas.read_table('out-cg-n14', sep=' *')
dfpy['code'] = 'py'
dfcg['code'] = 'cg'
df = pandas.concat((dfpy, dfcg))
df['s'] = df['s'].map(lambda x: round(x, 6))
df['h0'] = df['h0'].map(lambda x: round(x, 6))
df = df.set_index(['code', 'file', 'h0', 's'])
graphs = df.index.get_level_values('file').unique()
```

```python
dfx = df.xs(['data/nonsrg/A_G1.dat', 1.0], level=['file', 'h0']).unstack('code')

for meas in dfx.columns.get_level_values(0).unique():
    figure()
    dfx[meas].plot(style=['b-','ro'])
    ylabel(symbol(meas))
    savefig('compare_{}.pdf'.format(meas))
```

```python
dfx = df.xs(1.0, level='h0').unstack('code')

for ga, gb in combinations(graphs, 2):
    xa = dfx.xs(ga, level='file')
    xb = dfx.xs(gb, level='file')
    delta = 2 * (xa - xb).abs() / (xa + xb)
    delta['q2'].plot(style=['b-', 'ro'])
    ylabel('$\Delta(Q_2)$')
    legend(title='$h_0=1$')
    savefig('compare_delta_q2_h1.pdf')
```

```python
dfx = df.xs(0.95, level='s').unstack('code')

for ga, gb in combinations(graphs, 2):
    xa = dfx.xs(ga, level='file')
    xb = dfx.xs(gb, level='file')
    delta = 2 * (xa - xb).abs() / (xa + xb)
    delta['q2'].plot(style=['b-', 'ro'])
    xlabel('$h_0$')
    ylabel('$\Delta(Q_2)$')
    legend(title='$s=0.95$')
    savefig('compare_delta_q2_s0.95.pdf')
```
