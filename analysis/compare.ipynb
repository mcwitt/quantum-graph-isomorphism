```python
import analysis as a
import pandas as pd
import glob
import qgi
```

```python
data_cg = a.load_files(glob.glob('output/g13*.out'))
data_ap = pd.read_csv('output/n13-arpack.out', index_col=['ver', 'graph', 'h', 'dh', 's'])
data = pd.concat((data_cg, data_ap))
data.head()
```

```python
xs = data.xs(['g13a.txt', 1, 1e-3], level=['graph', 'h', 'dh']).unstack('ver')
for meas in a.measurements:
    figure()
    xs[meas, 'v0.3'].dropna().abs().plot(style='b-')
    xs[meas, 'ARPACK'].dropna().abs().plot(style='ro')
    yscale('log')
    ylabel('$|{}|$'.format(a.symbol(meas)))
    legend(loc='best', title='$h_0=10^{-3}$')
    #savefig('compare-{}-vs-s.pdf'.format(meas))
```

```python
xs = data.xs(['g13a.txt', 0.51, 1e-3], level=['graph', 's', 'dh']).unstack('ver')
for meas in a.measurements:
    figure()
    xs[meas, 'v0.3'].dropna().abs().plot(style='b-')
    xs[meas, 'ARPACK'].dropna().abs().plot(style='ro')
    xscale('log')
    ylabel('$|{}|$'.format(a.symbol(meas)))
    legend(loc='best', title='$s=0.51$')
    #savefig('compare-{}-vs-h.pdf'.format(meas))
```
