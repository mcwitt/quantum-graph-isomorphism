```python
import analysis as a
import pandas as pd
import glob
import qgi
import re
import scipy.sparse.linalg as arp
from itertools import combinations
```

# Input parameters

```python
graph_files = glob.glob('graphs/VMB-2013/g13*.txt')
s_vals = linspace(0.01, 0.99, 50)
h0vals = logspace(-3, 1, 49)
dh = 1e-3
```

# Run diagonalization

```python
def diagonalize(H, v0):
    energy, psi = eigsh(H, v0=v0, k=1, which='SA')
    return energy[0], psi.T[0]

graphs = [qgi.load_graph(g) for g in graph_files]
for g in graphs: assert len(g) == qgi.N
output = []

for graph, graph_file in zip(graphs, graph_files):
    psi0 = rand(D)
    for h0 in h0vals:
        for s in s_vals:
            h = h0 * np.ones(N)
            H = hamiltonian(graph, h, s)
            energy, psi0 = diagonalize(H, psi0)
            mz = mag_z(psi0)
            mx = mag_x(psi0)
            q2 = overlap(psi0)
            
            q2p = 0.
            
            for j in range(N):
                h[j] += 0.5 * dh
                H = hamiltonian(graph, h, s)
                energy, psi = diagonalize(H, psi0)
                fj = sigma_z(psi)
                
                h[j] -= dh
                H = hamiltonian(graph, h, s)
                energy, psi = diagonalize(H, psi0)
                fj = (fj - sigma_z(psi)) / dh
                q2p += np.sum(fj**2)
                
                h[j] += 0.5 * dh
                
            q2p = sqrt(q2p) / N
            graph_file = re.sub(r'.*/(.*)$', r'\1', graph_file)
            output.append((graph_file, h0, s, energy, mz, mx, q2, q2p))
            
df = pd.DataFrame.from_records(output,
    columns=['graph', 'h', 's', 'energy', 'mz', 'mx', 'q2', 'q2p'])

df['ver'] = 'ARPACK'
df['dh'] = dh
df = df.set_index(['ver', 'graph', 'h', 'dh', 's'])
df.to_csv('output/n13-arpack.out')
```
